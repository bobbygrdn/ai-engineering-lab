# VRAM Budgeting for Hugging Face Models
Estimate the GPU memory (GB) required for a model:

$$M \approx \frac{P \times 4}{Q} \times 1.2$$

- *P*: parameters in billions  
- *Q*: bits per weight (e.g., 16, 8)  
- *1.2*: 20 % overhead for KV cache/activations

```python
def vram_estimate(params_billion, bits_per_weight):
    """
    Estimate VRAM (GB) for a model.

    Args:
        params_billion (float): Parameter count in billions (P).
        bits_per_weight (int): Bit‑width per weight (Q).

    Returns:
        float: Estimated memory in GB.
    """
    # 4 bytes per 32‑bit float → 4 GB per billion parameters at 32‑bit
    base_gb = params_billion * (bits_per_weight / 8)   # convert bits→bytes
    return base_gb * 1.2
```

```python
models = {
    "Llama-3-8B-fp16":   {"P": 8, "Q": 16},
    "Llama-3-8B-Q8_0":   {"P": 8, "Q": 8},
    "Llama-3-8B-Q6_K":   {"P": 8, "Q": 6},
    "Llama-3-8B-Q4_0":   {"P": 8, "Q": 4},

    "Mistral-7B-fp16":   {"P": 7, "Q": 16},
    "Mistral-7B-Q8_0":   {"P": 7, "Q": 8},
    "Mistral-7B-Q6_K":   {"P": 7, "Q": 6},
    "Mistral-7B-Q4_0":   {"P": 7, "Q": 4},

    "Phi-3-mini-4B-fp16": {"P": 4, "Q": 16},
    "Phi-3-mini-4B-Q8_0": {"P": 4, "Q": 8},
    "Phi-3-mini-4B-Q6_K": {"P": 4, "Q": 6},
    "Phi-3-mini-4B-Q4_0": {"P": 4, "Q": 4},
}

for name, spec in models.items():
    mem = vram_estimate(spec["P"], spec["Q"])
    print(f"{name}: {mem:.2f} GB")
```

```python
gpu_vram_gb = 14

for name, spec in models.items():
    mem = vram_estimate(spec["P"], spec["Q"])
    fits = mem <= gpu_vram_gb
    print(f"{name} fits on GPU with {gpu_vram_gb}GB VRAM: {fits}")
```
